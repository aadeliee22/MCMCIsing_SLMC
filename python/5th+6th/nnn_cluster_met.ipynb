```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
J2 = 0.15
addpath = '/home/hyejin//' #'C:/Users/User/Desktop/'
path = addpath + 'mygit/MCM/python/txtfiles-6/-0.15~0.2'
cols = ['m', 'ms', 'e', 'sp_h']
cc = []
cc.append(pd.read_csv(path + '/cluster_1.txt', sep = ' ', header = 0)) #0
cc.append(pd.read_csv(path + '/cluster_1_eff.txt', sep = ' ', header = 0)) #1
cc.append(pd.read_csv(path + '/cluster_2.txt', sep = ' ', header = 0)) #2
mm = pd.read_csv(path + '/met_2.txt', sep = ' ', header = 0)
c = []
for i in range (len(cc)):
    c.append(cc[i][cc[i].J2==J2])
m = mm[mm.J2==J2]

print("J2 =", np.average(m.J2))
for j in range (4):
    plt.plot(m.temperature, m[cols[j]], 'r.', markersize = 10, markerfacecolor = None, linestyle = 'None', label = 'original')
    for i in range (len(c)):
        plt.plot(c[i].temperature, c[i][cols[j]], '.', markersize = 3, markerfacecolor = None, label = i)
    plt.ylabel(cols[j], fontsize=16)
    plt.xlabel("T", fontsize=16)
    plt.legend()
    plt.show()
```
